- Lets do this logic for all dataset

```python
def tokenize_and_aligen_labels(examples):
  # tokenize words into input ids
  tokenized_input =model_xlmr(examples['tokens'],is_split_into_words=True,truncation=True)


  labels =[]  # all tokens tags str with IGN tokens for special words and subword tokens that we will ignore during training
  for idx ,label in enumerate(examples['ner_tags']):

        # for each example

        word_ids =tokenized_input.word_ids(batch_index=idx)  # word_ids for each exaple

        previous_word_idx =None
        labels_ids =[]      # ids for each token and make -100 for sepecial tokesn and subwords tokens

        for word_idx in word_ids:
          if word_idx is None or word_idx ==previous_word_idx:  # special token or subword token that we need to mask during traing
              labels_ids.append(-100)
          elif word_idx !=previous_word_idx:
              labels_ids.append(label[word_idx])
          previous_word_idx =word_idx

        # store lables_ids after adding -100 id  for each example
        labels.append(labels_ids)

  # add all labels ids for all exaples tokenized input
  tokenized_input['labels'] =labels
  return tokenized_input

```

```python
# now lets encode dataset
def encode_panx_dataset(corpus):
  return corpus.map(tokenize_and_aligen_labels,batched=True,remove_columns=['langs', 'ner_tags', 'tokens'])
```

### FineTuning XLM-RoBERTa only on German
- make zero shot transfere in other languages for evaluation

```python
import numpy as

# make prediction suitable for f1_score metric to evalute performance
def align_predition(predictions ,label_ids):
    preds = np.argmax(predictions, axis=2)
    batch_size, seq_len = preds.shape
    labels_list, preds_list = [], []

    for batch_idx in range(batch_size):


        example_labels, example_preds = [], []
        for seq_idx in range(seq_len):
              # gnore label Ds = -100
              if label_ids[batch_idx, seq_idx] != -100:
                  example_labels.append(index2tag[label_ids[batch_idx][seq_idx]])
                  example_preds.append(index2tag[preds[batch_idx][seq_idx]])

        labels_list.append(example_labels)
        preds_list.append(example_preds)
    return preds_list, labels_list
```

- Define training arguments

```python
from transformers import TrainingArguments

num_epochs = 3
batch_size = 24
logging_steps = len(panx_de_encoded["train"]) // batch_size

model_name = f"{xlm_r_id}-finetuned-panx-de"

training_args = TrainingArguments(
                  output_dir=model_name,
                  log_level="error",
                  num_train_epochs=num_epochs,
                  per_device_train_batch_size=batch_size,
                  per_device_eval_batch_size=batch_size,
                  eval_strategy="epoch",
                  save_steps=1e6,
                  weight_decay=0.01,
                  disable_tqdm=False,
                  logging_steps=logging_steps,
                  push_to_hub=True
                  )
```

```python
import numpy as np
!pip install seqeval
from seqeval.metrics import f1_score

def compute_metrics(eval_pred):
  y_pred, y_true = align_predition(eval_pred.predictions,
  eval_pred.label_ids)
  return {"f1": f1_score(y_true, y_pred)}
```

```python
from transformers import DataCollatorForTokenClassification
from transformers import XLMRobertaForTokenClassification,AutoConfig
import torch

device =torch.device("cuda" if torch.cuda.is_available() else "cpu")

# set configiration
xlmr_config =AutoConfig.from_pretrained(xlm_r_id , num_labels=tags.num_classes,id2label=index2tag, label2id=tage2index)



data_collator =DataCollatorForTokenClassification(model_xlmr)

def model_init():
   return ( XLMRobertaForTokenClassification
    .from_pretrained(xlm_r_id, config=xlmr_config)
    .to(device))
```

```python
from transformers import Trainer

trainer = Trainer(model_init=model_init,
                  args=training_args,
                  data_collator=data_collator,
                  compute_metrics=compute_metrics,
                  train_dataset=panx_de_encoded["train"],
                  eval_dataset=panx_de_encoded["validation"],
                  tokenizer=model_xlmr
                  )
```

```python
trainer.train()
```

```python
# take any text and convert it ot tags

text_de = "Jeff Dean ist ein Informatiker bei Google in Kalifornien" # french text
def tag_text(text, tags, model, tokenizer):
  # et tokens with special characters
  tokens = tokenizer(text).tokens()
  # Encode the sequence into Ds
  input_ids = model_xlmr(text, return_tensors="pt").input_ids.to(device)
  # et predictions as distribution over 7 possible classes
  outputs = model(input_ids=input_ids)
  # Take argmax to get most likely class per token
  predictions = torch.argmax(outputs.logits, dim=2)
  # Convert to Datarame
  preds = [tags.names[p] for p in predictions[0].cpu().numpy()]
  return pd.DataFrame([tokens, preds], index=["Tokens", "Tags"])
tag_text(text_de, tags, trainer.model, model_xlmr)
```

### Fine tuning in multiple language at one
- finetuning on German and French at one

```python
from datasets import concatenate_datasets,DatasetDict

def concatenate_splits(corpus):
    keys =corpus[0].keys()
    data_after_concate =DatasetDict()

    for split in keys:
          data_after_concate[split] =concatenate_datasets([lang[split] for lang in corpus])

    return data_after_concate
```

```python
panx_de_fre =concatenate_splits([encode_panx_dataset(panx_ch['de']),encode_panx_dataset(panx_ch['fr'])])
```

```python
training_args.logging_steps = len(panx_de_fre["train"]) // batch_size
training_args.push_to_hub = True
training_args.output_dir = "xlm-roberta-base-finetuned-panx-de-fr"

trainer = Trainer(model_init=model_init,
                  args=training_args,
                  data_collator=data_collator,
                  compute_metrics=compute_metrics,
                  tokenizer=model_xlmr,
                  train_dataset=panx_de_fre["train"],
                  eval_dataset=panx_de_fre["validation"]
                  )
trainer.train()
trainer.push_to_hub(commit_message="Training completed!")
```
